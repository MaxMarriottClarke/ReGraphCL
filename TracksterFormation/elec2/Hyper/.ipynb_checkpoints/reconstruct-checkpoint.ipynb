```python
import torch
import torch_geometric
import uproot  # For loading ROOT files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score  # For purity evaluation, purity_score can be custom-defined

from torch_geometric.data import DataLoader 
from train import CCV1

from collections import defaultdict
```

```python
# Open the ROOT file
testpath = "/vols/cms/mm1221/Data/2e/test/"  # Replace with your ROOT file path
data_test = CCV1(testpath, max_events=12000, inp = 'test')

test_loader = DataLoader(data_test, batch_size=1, shuffle=False, follow_batch=['x_lc'])
```

```python
# Initialize model and load state dictionary
from train import Net
model = Net(hidden_dim=32, k_value=64, contrastive_dim=8)
checkpoint= torch.load('/vols/cms/mm1221/hgcal/CLe/Hyper/results/Layer3/results_lr0.001_bs32_hd32_k64_temp0.1_cd8/best_model.pt', map_location=torch.device('cpu'))
model.load_state_dict(checkpoint)  # Directly use the loaded checkpoint
model.eval()  # Set model to evaluation mode
```

```python
sim_vertices = data_test.stsCP_vertices_indexes
sim_x = data_test.stsCP_vertices_x
sim_y = data_test.stsCP_vertices_y
sim_z = data_test.stsCP_vertices_z
sim_energy = data_test.stsCP_vertices_energy
sim_centre_x = data_test.stsCP_barycenter_x
sim_centre_y = data_test.stsCP_barycenter_y
sim_centre_z = data_test.stsCP_barycenter_z
```

```python
# Initialize containers for predictions and matched simulated data
all_predictions = []
all_simulated_clusters = []
all_simulated_energies = []


for event_index, data in enumerate(test_loader):
    
    predictions_tuple = model(data.x_lc, data.x_lc_batch) 
    predictions = predictions_tuple[0]  
    predictions = predictions.detach().cpu().numpy()  
    all_predictions.append(predictions)


    simulated_clusters_vert = defaultdict(list)
    simulated_clusters_vert[0] = list(sim_vertices[event_index][0])  
    simulated_clusters_vert[1] = list(sim_vertices[event_index][1])  
    all_simulated_clusters.append(simulated_clusters_vert)

    simulated_energies = {
        0: sim_energy[event_index][0].tolist(),  
        1: sim_energy[event_index][1].tolist()   
    }
    all_simulated_energies.append(simulated_energies)


    total_sim_clusters_matched = 0
    total_sim_clusters = 0  
    all_scores = []  
```

```python
def calculate_distance(sim_centre_x, sim_centre_y, sim_centre_z):
    # Calculate pairwise distances between simulated cluster centers
    distances = []
    for event_idx in range(len(sim_centre_x)):
        x1, x2 = sim_centre_x[event_idx]
        y1, y2 = sim_centre_y[event_idx]
        z1, z2 = sim_centre_z[event_idx]
        distance = np.sqrt((x1 - x2)**2 + (y1 - y2)**2 + (z1 - z2)**2)
        distances.append(distance)
    return distances

def calculate_purity_vs_distance(
    all_predictions, 
    all_simulated_clusters, 
    all_simulated_energies, 
    sim_centre_x, 
    sim_centre_y, 
    sim_centre_z, 
    model_name, 
    threshold=0.2, 
    distance_bins=np.arange(0, 160, 5)
):
    # Calculate distances
    distances = calculate_distance(sim_centre_x, sim_centre_y, sim_centre_z)
    
    # Initialize results
    purity_per_bin = []
    error_per_bin = []  # To store binomial uncertainty
    bin_centers = []
    event_counts = []  # To store number of events in each bin

    total_matched_clusters = 0
    total_sim_clusters = 0  # For overall purity calculation

    for i in range(len(distance_bins) - 1):
        # Define the current bin range
        bin_min, bin_max = distance_bins[i], distance_bins[i + 1]
        bin_events = [idx for idx, d in enumerate(distances) if bin_min <= d < bin_max]

        # Skip bins with no events
        if not bin_events:
            purity_per_bin.append(0)
            error_per_bin.append(0)
            bin_centers.append((bin_min + bin_max) / 2)
            event_counts.append(0)
            continue

        # Calculate purity for events in the current bin
        bin_matched_clusters = 0
        bin_sim_clusters = 0

        for event_idx in bin_events:
            predictions = all_predictions[event_idx]

            # Step 1: Perform clustering for reconstructed clusters
            n_clusters = len(all_simulated_clusters[event_idx])  # Number of simulated clusters in this event
            kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0).fit(predictions)
            cluster_labels = kmeans.labels_

            # Organize reconstructed clusters
            predicted_clusters = defaultdict(list)
            for idx, label in enumerate(cluster_labels):
                predicted_clusters[label].append(idx)

            max_reco_index = max(max(cluster) for cluster in predicted_clusters.values())
            max_sim_index = max(max(cluster) for cluster in all_simulated_clusters[event_idx].values())
            max_index = max(max_reco_index, max_sim_index)

            # Initialize ordered_energies with the correct size
            ordered_energies = np.zeros(max_index + 1)

            # Populate ordered_energies with simulated cluster energies
            for cluster_id, vertices in all_simulated_clusters[event_idx].items():
                for vertex, energy in zip(vertices, all_simulated_energies[event_idx][cluster_id]):
                    ordered_energies[vertex] = energy

            # Match reconstructed clusters to simulated clusters
            matched_sim_clusters = set()
            for reco_cluster_id, reco_vertices in predicted_clusters.items():
                for sim_cluster_id, sim_vertices in all_simulated_clusters[event_idx].items():
                    shared_vertices = set(reco_vertices) & set(sim_vertices)
                    if not shared_vertices:
                        continue

                    total_reco_energy = sum(ordered_energies[v] for v in reco_vertices)
                    total_sim_energy = sum(ordered_energies[v] for v in sim_vertices)

                    fractions_reco = {v: ordered_energies[v] / total_reco_energy for v in shared_vertices}
                    fractions_sim = {v: ordered_energies[v] / total_sim_energy for v in shared_vertices}

                    numerator = 0
                    denominator = 0
                    for vertex in shared_vertices:
                        epsilon_k = ordered_energies[vertex]
                        term1 = (fractions_reco[vertex] - fractions_sim[vertex]) ** 2
                        term2 = fractions_sim[vertex] ** 2
                        numerator += min(term1, term2) * (epsilon_k ** 2)
                        denominator += term2 * (epsilon_k ** 2)

                    if denominator > 0:
                        score = numerator / denominator
                        if score < threshold:
                            matched_sim_clusters.add(sim_cluster_id)

            bin_matched_clusters += len(matched_sim_clusters)
            bin_sim_clusters += len(all_simulated_clusters[event_idx])

        total_matched_clusters += bin_matched_clusters
        total_sim_clusters += bin_sim_clusters

        # Calculate purity and error for this bin
        if bin_sim_clusters > 0:
            purity = bin_matched_clusters / bin_sim_clusters
            error = np.sqrt(purity * (1 - purity) / bin_sim_clusters)  # Binomial uncertainty
        else:
            purity = 0
            error = 0

        purity_per_bin.append(purity)
        error_per_bin.append(error)
        bin_centers.append((bin_min + bin_max) / 2)
        event_counts.append(len(bin_events))

    # Calculate overall purity
    overall_purity = total_matched_clusters / total_sim_clusters if total_sim_clusters > 0 else 0
    print(f"Overall Purity for {model_name}: {overall_purity:.4f}")

    # Save results for plotting later
    results = {
        "model_name": model_name,
        "bin_centers": bin_centers,
        "purity_per_bin": purity_per_bin,
        "error_per_bin": error_per_bin,
        "event_counts": event_counts,
        "overall_purity": overall_purity,
    }

    # Print the event counts for each bin
    print(f"Distance Bins and Event Counts for {model_name}:")
    for i, count in enumerate(event_counts):
        print(f"Bin {distance_bins[i]}-{distance_bins[i+1]} cm: {count} events")

    # Plot the purity vs distance with error bars
    plt.figure(figsize=(8, 6))
    plt.errorbar(bin_centers, purity_per_bin, yerr=error_per_bin, fmt='o', label=f'{model_name}', capsize=3)
    plt.plot(bin_centers, purity_per_bin, linestyle='-', label=f'{model_name} (line)')
    plt.xlabel('Distance [cm]')
    plt.ylim(0, 1)
    plt.ylabel('Purity')
    plt.title('Purity vs Distance')
    plt.grid()
    plt.legend()
    plt.show()

    return results
```

```python
def calculate_metrics_vs_distance(
    all_predictions, 
    all_simulated_clusters, 
    all_simulated_energies, 
    sim_centre_x, 
    sim_centre_y, 
    sim_centre_z, 
    model_name, 
    threshold=0.2, 
    energy_threshold=0.4, 
    distance_bins=np.arange(0, 20, 1)
):
    # Calculate distances
    distances = calculate_distance(sim_centre_x, sim_centre_y, sim_centre_z)

    purity_per_bin = []
    purity_error_per_bin = []
    merge_rate_per_bin = []
    merge_rate_error_per_bin = []
    efficiency_per_bin = []
    efficiency_error_per_bin = []
    bin_centers = []
    event_counts = []

    total_matched_clusters = 0
    total_sim_clusters = 0
    total_merged_clusters = 0
    total_reco_clusters = 0
    total_efficiency_clusters = 0

    for i in range(len(distance_bins) - 1):
        bin_min, bin_max = distance_bins[i], distance_bins[i + 1]
        bin_events = [idx for idx, d in enumerate(distances) if bin_min <= d < bin_max]

        # Skip bins with no events
        if not bin_events:
            purity_per_bin.append(0)
            purity_error_per_bin.append(0)
            merge_rate_per_bin.append(0)
            merge_rate_error_per_bin.append(0)
            efficiency_per_bin.append(0)
            efficiency_error_per_bin.append(0)
            bin_centers.append((bin_min + bin_max) / 2)
            event_counts.append(0)
            continue

        bin_matched_clusters = 0
        bin_sim_clusters = 0
        bin_merged_clusters = 0
        bin_reco_clusters = 0
        bin_efficiency_clusters = 0

        for event_idx in bin_events:
            predictions = all_predictions[event_idx]

            # Perform clustering for reconstructed clusters
            n_clusters = len(all_simulated_clusters[event_idx])
            kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0).fit(predictions)
            cluster_labels = kmeans.labels_

            predicted_clusters = defaultdict(list)
            for idx, label in enumerate(cluster_labels):
                predicted_clusters[label].append(idx)

            bin_reco_clusters += len(predicted_clusters)
            bin_sim_clusters += len(all_simulated_clusters[event_idx])

            max_reco_index = max(max(cluster) for cluster in predicted_clusters.values())
            max_sim_index = max(max(cluster) for cluster in all_simulated_clusters[event_idx].values())
            max_index = max(max_reco_index, max_sim_index)
            ordered_energies = np.zeros(max_index + 1)

            for cluster_id, vertices in all_simulated_clusters[event_idx].items():
                for vertex, energy in zip(vertices, all_simulated_energies[event_idx][cluster_id]):
                    ordered_energies[vertex] = energy

            matched_efficiency_clusters = set()
            matched_purity_clusters = set()

            for reco_cluster_id, reco_vertices in predicted_clusters.items():
                associated_sim_clusters = []
                for sim_cluster_id, sim_vertices in all_simulated_clusters[event_idx].items():
                    shared_vertices = set(reco_vertices) & set(sim_vertices)
                    if not shared_vertices:
                        continue

                    total_reco_energy = sum(ordered_energies[v] for v in reco_vertices)
                    total_sim_energy = sum(ordered_energies[v] for v in sim_vertices)

                    fractions_reco = {v: ordered_energies[v] / total_reco_energy for v in shared_vertices}
                    fractions_sim = {v: ordered_energies[v] / total_sim_energy for v in shared_vertices}

                    # Calculate Sim-to-Reco score
                    numerator = 0
                    denominator = 0
                    for vertex in shared_vertices:
                        epsilon_k = ordered_energies[vertex]
                        term1 = (fractions_reco[vertex] - fractions_sim[vertex]) ** 2
                        term2 = fractions_sim[vertex] ** 2
                        numerator += min(term1, term2) * (epsilon_k ** 2)
                        denominator += term2 * (epsilon_k ** 2)

                    if denominator > 0:
                        score = numerator / denominator
                        # Purity calculation (threshold-based)
                        if score < threshold:
                            matched_purity_clusters.add(sim_cluster_id)

                    # Efficiency calculation (energy threshold-based)
                    if total_sim_energy > 0:
                        energy_shared = sum(ordered_energies[v] for v in shared_vertices) / total_sim_energy
                        if energy_shared >= energy_threshold:
                            matched_efficiency_clusters.add(sim_cluster_id)
                            associated_sim_clusters.append(sim_cluster_id)

                # Check for merging
                if len(associated_sim_clusters) > 1:
                    bin_merged_clusters += 1

            bin_efficiency_clusters += len(matched_efficiency_clusters)
            bin_matched_clusters += len(matched_purity_clusters)

        total_matched_clusters += bin_matched_clusters
        total_sim_clusters += bin_sim_clusters
        total_merged_clusters += bin_merged_clusters
        total_reco_clusters += bin_reco_clusters
        total_efficiency_clusters += bin_efficiency_clusters

        # Purity and its error
        if bin_sim_clusters > 0:
            purity = bin_matched_clusters / bin_sim_clusters
            purity_error = np.sqrt(purity * (1 - purity) / bin_sim_clusters)
        else:
            purity = 0
            purity_error = 0

        # Merge rate and its error
        if bin_reco_clusters > 0:
            merge_rate = bin_merged_clusters / bin_reco_clusters
            merge_rate_error = np.sqrt(merge_rate * (1 - merge_rate) / bin_reco_clusters)
        else:
            merge_rate = 0
            merge_rate_error = 0

        # Efficiency and its error
        if bin_sim_clusters > 0:
            efficiency = bin_efficiency_clusters / bin_sim_clusters
            efficiency_error = np.sqrt(efficiency * (1 - efficiency) / bin_sim_clusters)
        else:
            efficiency = 0
            efficiency_error = 0

        purity_per_bin.append(purity)
        purity_error_per_bin.append(purity_error)
        merge_rate_per_bin.append(merge_rate)
        merge_rate_error_per_bin.append(merge_rate_error)
        efficiency_per_bin.append(efficiency)
        efficiency_error_per_bin.append(efficiency_error)
        bin_centers.append((bin_min + bin_max) / 2)
        event_counts.append(len(bin_events))

    # Overall metrics
    overall_purity = total_matched_clusters / total_sim_clusters if total_sim_clusters > 0 else 0
    overall_merge_rate = total_merged_clusters / total_reco_clusters if total_reco_clusters > 0 else 0
    overall_efficiency = total_efficiency_clusters / total_sim_clusters if total_sim_clusters > 0 else 0

    print(f"Overall Purity for {model_name}: {overall_purity:.4f}")
    print(f"Overall Merge Rate for {model_name}: {overall_merge_rate:.4f}")
    print(f"Overall Efficiency for {model_name}: {overall_efficiency:.4f}")

    # Plot results
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.errorbar(bin_centers, purity_per_bin, yerr=purity_error_per_bin, fmt='o', label='Purity', capsize=3)
    plt.xlabel('Distance [cm]')
    plt.ylabel('Purity')
    plt.title('Purity vs Distance')
    plt.grid()
    plt.legend()

    plt.subplot(1, 3, 2)
    plt.errorbar(bin_centers, merge_rate_per_bin, yerr=merge_rate_error_per_bin, fmt='o', label='Merge Rate', capsize=3)
    plt.xlabel('Distance [cm]')
    plt.ylabel('Merge Rate')
    plt.title('Merge Rate vs Distance')
    plt.grid()
    plt.legend()

    plt.subplot(1, 3, 3)
    plt.errorbar(bin_centers, efficiency_per_bin, yerr=efficiency_error_per_bin, fmt='o', label='Efficiency', capsize=3)
    plt.xlabel('Distance [cm]')
    plt.ylabel('Efficiency')
    plt.title('Efficiency vs Distance')
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()

    return {
        "purity_per_bin": purity_per_bin,
        "merge_rate_per_bin": merge_rate_per_bin,
        "efficiency_per_bin": efficiency_per_bin,
        "overall_purity": overall_purity,
        "overall_merge_rate": overall_merge_rate,
        "overall_efficiency": overall_efficiency,
    }
```

```python
model_results = []
model_results.append(calculate_metrics_vs_distance(
    all_predictions, 
    all_simulated_clusters, 
    all_simulated_energies, 
    sim_centre_x, 
    sim_centre_y, 
    sim_centre_z,  
    model_name="Model_DynN"))

```

```python
import pandas as pd

# Flatten the model_results dictionary into a list of rows for each bin
csv_rows = []
for result in model_results:
    model_name = result['model_name']
    bin_centers = result['bin_centers']
    purity_per_bin = result['purity_per_bin']
    error_per_bin = result['error_per_bin']
    event_counts = result['event_counts']
    overall_purity = result['overall_purity']

    for bin_center, purity, error, count in zip(bin_centers, purity_per_bin, error_per_bin, event_counts):
        csv_rows.append({
            "Model": model_name,
            "Bin_Center": bin_center,
            "Purity": purity,
            "Error": error,
            "Event_Count": count,
            "Overall_Purity": overall_purity
        })

# Create a DataFrame from the rows
df = pd.DataFrame(csv_rows)

# Save the DataFrame to a CSV file
output_csv_path = "DY_results.csv"
df.to_csv(output_csv_path, index=False)

print(f"Results successfully saved to {output_csv_path}")
```
